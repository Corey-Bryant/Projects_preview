## This notebook is an example of an ANOVA conducted using R

```r
# Getting data

datafilename="http://personality-project.org/r/datasets/maps.mixx.epi.bfi.data"

person.data =read.table(datafilename,header=TRUE)
```

```r
# Getting a summary of the data

library(psych)

describe(person.data)
```

## Showing some simple graphing examples (see below)

```r
# Box plotting just the first 5 variables

boxplot(person.data[1:5], main= "Boxplot of the first 5 Variables")
```

```r
# Creating a histogram of the extraversion scores

hist(person.data$epiE, main = "Extraversion Scores")
```

```r
# Creating a scatter plot

plot(person.data$epiE, person.data$bfext,
     main= "Comparing Measures of Extraversion")
```

```r
pairs.panels(person.data[1:5], 
             main = "Scatter Plot, Histogram, and Correlation Matrix Plot")
```

## Correlations

```r
# Showing correlation matrix rounding to two decimal places

round(cor(person.data, use = "pairwise"), 2)
```

## Regression

From the correlation matrix conducted, we know Neuroticism and Anxiety are related to depression. Let's conduct a regression analysis with this as predictors of depression.

```r
model1 = lm(bdi ~ epiNeur, data= person.data)
summary(model1)
```

```r
model2 = lm(bdi~ epiNeur + traitanx, data = person.data)
summary(model2)
```

Looking at the multiple R-squared value, the second model accounted for more variance. This indicates that our second model is a better predictor of depression.
